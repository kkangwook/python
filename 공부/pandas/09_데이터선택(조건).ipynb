# 9.데이터 선택(조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키']>=185 #T F로 나타냄
```

```python
filt = (df['키'] >= 185) #키 185이상인 애들의 모든 정보 가져옴
df[filt]
```

```python
df[df['키']>=185] #이렇게도 가능
```

```python
df[df['키']>185]['수학']  #이렇게도 가능
```

```python
df[~filt] #False 값인 애들만 적용->키 185 미만인 애들만 출력
```

```python
df[~(df['키']>=185)] #이렇게도 가능, ()필요
```

## df.loc['row','col'] 사용하여 필터링

```python
df.loc[df['키'] >= 185, '수학']  # 키가 185 이상인 학생들의 수학 데이터
```

```python
df.loc[df['키'] >= 185, ['이름', '수학', '과학']] # 키가 185 이상인 학새들의 여러 열 가져오기
```

## 다양한 조건

### 그리고 & and

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고')] # 키가 185 이상 & 북산고 학생 데이터, ()로 감싸줘야함
```

### 또는 | or

```python
df.loc[(df['키'] < 170) | (df['키'] > 200)] # 키가 170 보다 작거나 or 200 보다 큰 학생 데이터
```

### str함수
문자열 처리, 비교

```python
filt=df['이름'].str.startswith('송') #송으로 시작하는 문자열을 가진 사람을 뽑음
print(df['이름'].str)  #안보임
print(filt) #T,F로 나오고 이를 df에 넣으면 row로 출력
df[filt]
```

```python
filt = df['이름'].str.contains('태') # 이름에 '태'가 들어가는 사람
print(filt)
df[filt]
```

```python
df[~filt] # 이름에 '태'가 들어가는 사람을 제외
```

```python
langs = ["Python", "Java"]
filt = df['SW특기'].isin(langs) # .isin()함수=데이터 중에 ()안에 들어간 애들만 뽑기
df[filt] #대문자는 안뽑힘
```

```python
langs = ["python", "java"]
filt = df['SW특기'].str.lower().isin(langs)  #데이터를 소문자화 한 다음 langs안에 있는 애들만 출력
df[filt]
```

```python
#자바와 자바스크립트 둘다 같이 뽑기
filt = df['SW특기'].str.contains('Java')
df[filt] #하지만 에러
```

```python
filt = df['SW특기'].str.contains('Java')
filt #NaN값이 나왔기 때문에 filtering이 안된거였음
```

```python
df['SW특기'].str.contains('Java', na=True) # NaN 데이터에 대해서 True 로 설정
```

```python
df['SW특기'].str.contains('Java', na=False) # NaN 데이터에 대해서 False 로 설정
```

```python
filt = df['SW특기'].str.contains('Java', na=False) #이렇게 해야 가져올 수 있음
df[filt]
```
