# 4. 파일 저장 및 열기
dataframe객체를 excel, csv, txt등 형태의 파일로 저장 및 열기

```python
import pandas as pd
data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
df = pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번', '6번', '7번', '8번'])
df.index.name = '지원번호'
df
```

## 저장하기
### csv 파일로 저장
위치는 c:/사용자->깽욱->pythondataworkspace->공부->pandas  
오류 없게 하기 위해 encoding='utf-8-sig'필요

```python
df.to_csv('score.csv',encoding='utf-8-sig') # to_csv함수 사용->주피터 노트북에 파일 생성,내 컴터에도 생성
#맨 앞의 정보는 만들 파일의 이름
```

```python
df.to_csv('score.csv',encoding='utf-8-sig',index=False) #지원번호란 인덱스 없이 저장
```

### 텍스트(.txt)파일로 저장
그대로 .to_csv사용하면 됨

```python
df.to_csv('score.txt') #그냥 하면 안띄어쓰고 생성
```

```python
df.to_csv('score.txt',sep='\t') #각 칼럼을 tab으로 나눔
```

### 엑셀파일로 저장
.to_excel함수 사용, 확장자는 .xlsx로

```python
df.to_excel('score.xlsx')
```

## 열기-import pandas 필요
### csv 파일 열기
.read_csv('이름')함수 사용

```python
df=pd.read_csv('score.csv')
df
```

```python
df = pd.read_csv('score.csv', skiprows=1) # 지정된 갯수 만큼의 row 를 건너뜀
df #맨윗줄 무시하고 그다음부터 가져와 채치수 줄이 컬럼 이름이 됨
```

```python
df = pd.read_csv('score.csv', skiprows=[1, 3, 5]) # 리스트 형태로-> 1, 3, 5 row 는 제외 (row 0부터 시작)
df #채치수, 송태섭, 강백호 줄은 제외
```

```python
df = pd.read_csv('score.csv', nrows=4) # 지정된 갯수 만큼의 row 만 가져옴
df
```

```python
df = pd.read_csv('score.csv', skiprows=2, nrows=4) # 처음 2 row 무시, 이후에 4 row 를 가져옴
df
```

### 텍스트(.txt)파일 열기
똑같이 .read_csv함수 사용, sep사용해야 함

```python
df = pd.read_csv('score.txt', sep='\t')
df #지원번호가 컬럼이 되었음->set_index함수 사용하거나 밑의 옵션 사용
```

```python
df = pd.read_csv('score.txt', sep='\t', index_col='지원번호') #지원번호를 인덱스로
df
```

### 엑셀 파일 열기
.read_excel함수 사용

```python
df = pd.read_excel('score.xlsx')
df #지원번호의 칼럼화되어있음
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호') #컬럼을 지정해 인덱스로
df
```
