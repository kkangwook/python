# 12.데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

## column수정
일부를 수정할때-> df['컬럼명'].replace({'a':'b'},inplace=True): dic의 형태, a->b로  
줄 전부를 수정할때는 df['컬러명']=바꾸고싶은 값

```python
df['학교'].replace({'북산고':'상북고','능남고':'무슨고'})
```

```python
df #변화없음
```

```python
df['학교'].replace({'북산고':'상북고'},inplace=True)
df
```

```python
# 만약 대소문자 차이가 있어 replace를 바로 못쓸때는 데이터를 한 유형으로 몰아줌
df['SW특기'] = df['SW특기'].str.lower() #대문자 하고 싶으면 upper()하면 됨
df
```

```python
df['학교'] = df['학교'] + '등학교' # 학교 데이터 + 등학교
df
```

## column 추가
R의 df$new=원하는값 하는 식으로 df['새로운열이름']=원하는값 하면 됨

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회'] #새로운 '총합' 열 추가
df
```

```python
df['결과'] = 'Fail' # 결과 Column 을 추가하고 전체 데이터는 Fail 로 초기화
df
```

```python
#총합이 400 이상이면 결과열을 pass로
df.loc[df['총합']>400,'결과']='pass'
df
```

## column삭제
.drop(columns='삭제하고싶은 열이름',inplace=True)

```python
df.drop(columns='총합') #총합 column을 삭제
```

```python
df #반영은 안됨->반영시키고 싶으면 inplace=True로
```

```python
df.drop(columns=['국어', '영어', '수학']) # 국어, 영어, 수학 Column 을 삭제
```

## Row 삭제
drop에서 column대신 index(row)값을 입력

```python
df.drop(index='4번') # 4번 학생 데이터 row 를 삭제
```

```python
#수학 80점 밑인 애들 지우기
#index=''안에 수학80안되는 애들의 인덱스 값을 가져오면 됨
df.drop(index=df[df['수학']<80].index)
```

```python
df #반영안되므로 inplace=True해줘야함
```

## Row 추가

## Cell 수정

```python
#서태웅의 sw특기를 파이썬으로 수정해보기
df.loc['4번','SW특기']='python'
df
```

```python
#여러개 바꾸기
#강백호의 학교는 능남고등학교, SW특기는 C로 변경하기
df.loc['5번', ['학교', 'SW특기']] = ['능남고등학교', 'C'] 
df
```

```python
#여러개 바꿀때
df.loc[['1번','3번'],['이름','키']]=[['aa',1],['bb',2]]
df
```

## Column 순서 변경

```python
#데이터프레임의 일부분을 가져올때->
df[['이름','학교','키','SW특기']] #단 안에있는 칼럼들은 실제 df안에 있어야함
```

```python
cols = list(df.columns)
cols
```

```python
# 맨 뒤에 있는 결과 Column 을 앞으로 가져오고, 나머지 Column 들과 합쳐서 순서 변경
#이때 cols[-1]은 값이므로 리스트를 씌워줘서 뒤의 cols[0:-1]의 리스트와 합체 가능하게끔 함
df = df[[cols[-1]] + cols[0:-1]] 
df
```

## 컬럼 시프트
데이터를 아래로 하나씩 이동

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키_시프트']=df['키'].shift(1) #몇칸 아래로 이동시킬지
df
```

```python
df['키_시프트']=df['키'].shift(-1) #-하면 위로 이동
df
```
