# 13.함수적용

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['학교'] = df['학교'] + '등학교' #간단한 것은 가능
df
```

```python
# 키열에 cm붙이기
df['키']=df['키']+'cm'
df #에러남 ->int와 str은 더할수 없어서
```

## 데이터에 함수 적용(apply함수)
[여러데이터의리스트].apply(fun)하면 리스트 안의 요소들 하나하나 뽑아내 함수 적용시키고 반환

```python
#df['키']=str(df['키'])+'cm'->이것도 불가
#각각의 데이터에 함수 적용시키는 apply필요
def add_cm(height):
    return str(height)+'cm'

df['키']=df['키'].apply(add_cm)
df
```

```python
#SW특기의 첫 문자는 대문자, 나머지는 소문자로 변경하기
def cap(lang):
    if pd.notnull(lang): #NaN이면 False, NaN이 아니면 True반환
        return lang.capitalize() #첫글자는 대문자로, 나머지는 소문자로로 만드는 함수
    return lang  #NaN은 그대로

df['SW특기']=df['SW특기'].apply(cap)
df
```
