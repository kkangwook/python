# Matplotlib
다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

## 1.그래프기본

```python
x=[1,2,3] #x값
y=[2,4,8] #y값
plt.plot(x,y) #plot함수
```

```python
plt.plot(x,y)
plt.show() #추가텍스트정보없이 그래프만 보이게끔
```

## Title 설정

```python
plt.plot(x,y)
plt.title('line graph')
```

```python
plt.plot(x, y)
plt.title('꺾은선 그래프') #한글로 하면 오류나옴
```

## 한글 폰트 설정

```python
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic' #windows에서 폰트 설정
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac에서는 이걸로
matplotlib.rcParams['font.size'] = 15 # 글자 크기
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용 가능한 폰트 확인
[f.name for f in fm.fontManager.ttflist] #위 목록들 중 이름만 뽑아오기
```

```python
plt.plot(x, y)
plt.title('꺾은선 그래프') #이제 한국어 나옴
```

```python
plt.plot([-1,0,1],[-5,-1,2]) #x,y값 직접넣기->이때 축 부분에 -부분이 깨짐
```

```python
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
plt.plot([-1,0,1],[-5,-1,2]) #-잘 나옴
```

```python
#따라서 matplotlib사용전에 항상 이거 런 하고 사용
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```
