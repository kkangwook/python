# 8. 막대그래프 기본

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] # 키
colors = ['r', 'g', 'b']

plt.bar(labels, values,color=colors, alpha=0.5,width=0.5) #bar=막대그래프 함수, labels=x축, values=y축, color=색, alpha=투명도, width=막대두께
#color=['r','g','b'] 이런식으로 넣어도 됨
plt.ylim(175, 195) #y 범위
```

```python
plt.bar(labels, values, width=0.3)
plt.xticks(rotation=45) # x 축의 이름 데이터 각도를 45 도로 설정
plt.yticks(rotation=45) # y 축의 키 데이터 각도를 45도로 설정
```

## 직접 이름 설정하기

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] # 키
ticks = ['1번학생', '2번학생', '3번학생']

plt.bar(labels, values)
plt.xticks(labels, ticks, rotation=90) #실제이름->내가 바꿀 이름, 각도
```
