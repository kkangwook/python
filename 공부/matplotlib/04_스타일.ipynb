# 4. 스타일
추가적인 세부적인 내용들  
1.점-점 표시법, 종류, 크기, 테두리 색, 안쪽 색  
2.선-선종류, 두께  
3.선-선 색깔  
4.위에 꺼를 한꺼번에 나타내는 법  
5.축약어  
6.투명도  
7.그래프 자체의 크기와 배경색

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y, linewidth=5) #선의 두께
```

## marker

```python
plt.plot(x, y, marker='o', markersize=20, markeredgecolor='red', markerfacecolor='yellow') #마커 종류, 크기, 테두리 섹, 안쪽 색
#구글에 matplotlib marker치면 가능한 마커 종류들 알려줌 ex) v, x등
```

```python
plt.plot(x, y, marker='o', linestyle='None') #라인 없이 점만
```

```python
plt.plot(x, y, marker='v',markersize=20) #마커 사이즈
```

## 선 스타일

```python
plt.plot(x, y, linestyle=':') #선 종류-> :,--,-.,-
#구글에 matplotlib linestyle치면 여러 종류 나옴
```

## 선 색깔

```python
plt.plot(x, y, color='pink') #색-ex) pink, #ff0000(rgb값 직접 설정-구글에 color picker치면 됨), g, b등
#구글에 matplotlib color치면 나옴
```

## 포맷

```python
plt.plot(x, y, 'ro--') # color(red), marker('o'), linestyle(--)
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, 'go') #linestyle은 none이어서 선 안나옴
```

## 축약어
구글에 matplotlib markerfacecolor mfc라 치고 맨 위에 사이트 들어가면 축약어 나옴

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='blue', ls=':')
#mfc=marker face color, ms=marker size, mec=marker edge color, ls=line style
```

## 투명도

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, alpha=0.2) # alpha : 투명도 (0~1)
```

## 그래프 크기

```python
plt.figure(figsize=(10, 5)) #x축과 y축의 비율
plt.plot(x, y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x, y)
```

```python
plt.figure(figsize=(10, 5), dpi=150) # dots per inch, 그래프의 해상도(크기)
plt.plot(x, y)
```

## 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x, y)
```
