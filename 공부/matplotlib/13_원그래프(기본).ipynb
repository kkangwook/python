# 13. 원그래프(기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
values = [30, 25, 20, 13, 10, 2] #선호하는 언어들 수
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC'] #선호하는 언어

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False)
# value=넓이 비중, label=원 바깥 이름, autopct(autopercent)=비율표시(%.1f:비율을 소수점 첫째 자리 까지, %%:%), startangle=90도, counterclock=false시계방향으로
plt.show()
```

```python
#쉽게 알아보게 간격 뛰우기
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
# explode = [0.2, 0.1, 0, 0, 0, 0]-> 얘로 하면 일일히 다르게 설정 가능
explode = [0.05] * 6    #0.05 6번 반복

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
#범례 만들기
plt.pie(values, labels=labels, explode=explode)
plt.title('언어별 선호도')
plt.legend(loc=(1.2, 0.3), title='언어별 선호도') #loc은 중심으로 부터의 거리, legend에도 제목 넣을 수 있음
plt.show()
```
