# 15. 산점도 그래프 scatter plot
두 변수의 상관관계 보여줌

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
import pandas as pd
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
plt.scatter(df['영어'], df['수학']) #x축=영어, y축=수학
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
#점 크기 랜덤으로
import numpy as np
sizes = np.random.rand(8) * 1000 #random.rand(x) 0~1사이의 값 x개 출력
sizes
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes) #s=점들 크기
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
sizes = df['학년'] * 500 # 1학년 = 500, 2학년 = 1000, 3학년 = 1500
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
# c:색깔 분포(학년에 따라 다른색), cmap: 색 종류, alpha=투명도
#구글 matplotlib cmap치면 종류나옴->그라데이션으로 나오면 알아서 색이 진행됨에 따라 정해줌
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
#그래프 밖에 컬러바 표시
plt.colorbar(ticks=[1, 2, 3], label='학년', shrink=0.5, orientation='horizontal') 
#ticks:표현될 범위, label: 컬러바의 이름, shrink: 컬러바 크기(0~1), orientation: horizontal or vertical
```
