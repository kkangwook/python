# 02. 축
각 축의 이름, 색, 위치, 범위 설정 방법

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
plt.title('꺾은선 그래프', fontdict={'family':'HYGungSo-Bold', 'size':20}) # 설정된 폰트 외의 이번만 개별 폰트, 크기 설정
```

```python
plt.plot(x, y)
plt.xlabel('X축', color='red',loc='right',size=100) #x축 이름과 색 위치(left,center,rigth)
plt.ylabel('Y축', color='#00aa00',loc='top') #y축 이름과 색(RGB중 G에만 값을 주고 나머지는 0) 위치(top,center,bottom)
```

```python
plt.plot(x, y)
plt.xticks([1, 2, 3]) #x축 범위
plt.yticks([3, 6, 9, 12]) #y축 범위
plt.show()
```
